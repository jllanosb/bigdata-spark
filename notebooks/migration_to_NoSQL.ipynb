IntegraciÃ³n con MongoDB

1. Instalar la extensiÃ³n MongoDB for VS Code.
2. Conectarse a la base de datos local mongodb://localhost:27017.
3. Crear la base de datos retail_spark y la colecciÃ³n prediction.
4. Insertar los registros del archivo limpio (transaccion_enriquecida.csv) a la colecciÃ³n prediction usando Python:

```python
import pandas as pd
from pymongo import MongoClient

# pip install dotenv
import os
from dotenv import load_dotenv

load_dotenv()

MONGO_URI = os.getenv("MONGODB_URI") # Localhost
client = MongoClient(MONGO_URI)
db = client["retail_spark"]
collection = db["prediction"]

"""
MONGO_URI_ATLAS = os.getenv("MONGODB_URI_ATLAS") # MongoDB Atlas
client_atlas = MongoClient(MONGO_URI_ATLAS)
db_atlas = client_atlas["retail_spark"]
collection = db_atlas["prediction"]
"""

df = pd.read_csv('dataset/empleado.csv')
data = df.to_dict(orient="records")
collection.insert_many(data)
print("Datos insertados en MongoDB correctamente.")
```

## âœ… Mostrar todos los documentos

ðŸ”¹ Imprimir cada documento completo (cada fila del CSV convertido a JSON) almacenado en la colecciÃ³n prestamos.

```python
print("Visualizar datos insertados en MongoDB.")
for doc in collection.find():
    print(doc)
```
